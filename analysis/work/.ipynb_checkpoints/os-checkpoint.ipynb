```python
import numpy as np
import pandas as pd
import json
import re
from operator import add
from ua_parser import user_agent_parser
from IPython.display import display

def os_filter(text):
    parsed_string = user_agent_parser.ParseOS(text)
    family = parsed_string['family']
    return family

def browser_filter(text):
    parsed_string = user_agent_parser.ParseUserAgent(text)
    family = parsed_string['family']
    return family

def os_estimator(text):
    regex = r'\"Math.tan\(-1e300\)\":([\-\.0-9]+)'
    pattern = re.compile(regex)
    match = pattern.search(text)
    value = float(match.group(1))
    if value == -4.987183803371:
        return "Windows"
    elif value == -1.4214488238747:
        return "Linux 64"
    elif value == 0.8831488831618:
        return "Linux 32"
    else:
        return "macOS"

df = pd.read_json('data.json')

display(df)

os = list(df.apply((lambda x:
                   os_filter(x.userAgent)
                 ), axis=1))
display(os)

browser = list(df.apply((lambda x:
                   browser_filter(x.userAgent)
                 ), axis=1))
display(browser)

estimate_os = list(df.apply((lambda x:
                                     os_estimator(x.math)), axis=1))
display(estimate_os)
```
