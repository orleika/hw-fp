```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

n = 50; N = 1000

x = np.linspace(-3, 3, n)
X = np.linspace(-3, 3, N)

pix = np.pi * x
y = np.sin(pix) / pix + 0.1 * x + 0.2 * np.random.randn(n)

x = x.reshape(-1, 1)
X = X.reshape(-1, 1)
y = y.reshape(-1, 1)

plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression

clf = LinearRegression()
clf.fit(x, y)
```

```python

p = clf.predict(X)

plt.scatter(x, y)
plt.plot(X,p)

print(clf.score(x, y))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression
from mpl_toolkits.mplot3d import Axes3D
from IPython.display import display

#Boston HOuse Pricesのデータをロード
boston = load_boston()
boston_df = pd.DataFrame(boston.data)
boston_df.columns = boston.feature_names
boston['Price'] = boston.target #目的変数をデータフレームに追加
```

```python
df = pd.DataFrame()
df['RM'] = boston_df['RM']
df['LSTAT'] = boston_df['LSTAT']

X_multi = df
Y_target = boston.target

display(X_multi)
display(Y_target)
```

```python
#モデル生成とフィッティング
lreg = LinearRegression()
```

```python
lreg.fit(X_multi, Y_target)
a1, a2 = lreg.coef_ #係数
b = lreg.intercept_ #切片
```

```python
x, y, z = np.array(df['RM']), np.array(df['LSTAT']), np.array(Y_target)
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter3D(np.ravel(x), np.ravel(y), np.ravel(z), c = 'red')

#3D描画（回帰平面の描画）
X, Y = np.meshgrid(np.arange(0, 10, 1), np.arange(0, 40, 1))
Z = a1 * X + a2 * Y + b
ax.plot_surface(X, Y, Z, alpha = 0.5) #alphaで透明度を指定
ax.set_xlabel("RM")
ax.set_ylabel("LSTAT")
ax.set_zlabel("Price")

plt.show()
```

```python
import numpy as np
import pandas as pd
import json
import re
from IPython.display import display
from sklearn.linear_model import LinearRegression
df = pd.read_json('data.json')
memory = df.apply((lambda x:
                  json.loads(x.memory)
                 ), axis=1)
X = np.array(list(memory.apply((lambda x:
                 list(np.array(x) / x[0])
             ))))

def memory_speed_filter(text):
    regex = r'type 17.+ ([0-9]+) (MH|MT)'
    pattern = re.compile(regex)
    match = pattern.search(text.replace('\n', ''))
    return int(match.group(1))

def cpu_speed_filter(text):
    regex = r'type 4.+Current Speed: ([0-9]+) MH'
    pattern = re.compile(regex)
    match = pattern.search(text.replace('\n', ''))
    return int(match.group(1))
    
speed = df.apply((lambda x:
                  [memory_speed_filter(x.smbios), cpu_speed_filter(x.smbios)]
                 ), axis=1)
Y = np.array(list(speed))

display(X)
display(Y)
```

```python
clf = LinearRegression()
clf.fit(X, Y)
display(clf.coef_)
display(clf.intercept_)
display(clf.score(X, Y))
```
