```python
import numpy as np
import pandas as pd
import json
import re
from operator import add
from ua_parser import user_agent_parser
from IPython.display import display
from sklearn.linear_model import LinearRegression
import sklearn.svm as svm
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
from plotly.graph_objs import Scatter, Figure, Layout, Bar, Margin
init_notebook_mode(connected=True)

def cpu_core_filter(text):
    regex = r'Core Count: ([0-9])'
    pattern = re.compile(regex)
    match = pattern.search(text.replace('\n', ''))
    if match == None:
        return 0
    return int(match.group(1))

def cpu_thread_filter(text):
    regex = r'Thread Count: ([0-9])'
    pattern = re.compile(regex)
    match = pattern.search(text.replace('\n', ''))
    if match == None:
        return 0
    return int(match.group(1))

df = pd.read_json('data.json')
# display(df)

cpu = np.array(list(df.apply((lambda x:
                 [cpu_core_filter(x.smbios), cpu_thread_filter(x.smbios)]
                 ), axis=1)))
display(cpu)

worker = df.apply((lambda x:
                  json.loads(x.worker)
                 ), axis=1)

X = np.array(list(worker.apply((lambda x:
                 list(np.array(x) / x[0])
             ))))

display(X)

Y = cpu
```

```python
trim_sample = [6, 20, 29, 37, 41, 42]
X_trimmed = np.delete(X, trim_sample, 0)
Y_trimmed = np.delete(Y, trim_sample, 0)
display(X_trimmed)
display(Y_trimmed)
```

```python
clf = LinearRegression()
clf.fit(X, Y)
display(clf.coef_)
display(clf.intercept_)
display(clf.score(X_trimmed, Y_trimmed))
```
